```python
import pandas as pd
import tensorflow as tf
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# open dataset
data = pd.read_csv("concrete_data.csv")
data.head()
```

```python
# segement dataset into X and Y features
data_x = data[data.columns[:-1]]
data_y = data['Strength']
```

```python
# perform the data split
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.3, shuffle=True, random_state=10)
```

```python
# create model
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=data_x.shape[1]),
    tf.keras.layers.Dense(10, activation=tf.keras.activations.relu),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss=tf.keras.losses.mean_absolute_error)
```

```python
# train model
model.fit(train_x, train_y, epochs=50)
```

```python
# evaluate MAE of the model
res = model.predict(test_x)
MAE = mean_absolute_error(test_y, res)
MAE
```

```python
# iterating the training and evaluation stage 50 times
ind = 0
MAE = []
while ind < 50:

    # step 1
    train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.3, shuffle=True, random_state=10)

    # step 2
    model.fit(train_x, train_y, epochs=50)

    # step 3
    res = model.predict(test_x)
    mae = mean_absolute_error(test_y, res)

    # update MAE
    MAE.append(mae)

    ind+= 1
```

```python
# mean and standard deviation
mean = np.mean(MAE)
std = np.std(MAE)

print(f"Mean : {mean} \nStandardDeviation : {std}")
```
